```python
from nltk.corpus.reader.plaintext import PlaintextCorpusReader
corpus = PlaintextCorpusReader('C:\\Users\\Shreyaa','sample.txt')
#extracting the file ids within the corpus
print(corpus.fileids())
#extracting paragraphs in the text
paragraph = corpus.paras()
print("Total Paragraphs:", len(paragraph))
#extracting sentences
sentences = corpus.sents()
print("Total Sentences:",len(sentences))
print("First sentence:",sentences[0])
#extract words from the corpus
print("Words in the corpus:",corpus.words())
freq = nltk.FreqDist(corpus.words())
#common words 
print("Common Words along with frequency:", freq.most_common(10))
#specific words 
print("Number of occurences of specific Word - computer: ", freq.get("computer"))

```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
fm=open("sample.txt",'r',errors='ignore')
x=fm.read()
print(x)
porter = PorterStemmer()
tokens = word_tokenize(x)
print("\n{0:20}{1:20}\n".format("Word","Stem"))
for t in tokens:
     stem_word = porter.stem(t)
     print ("{0:20}{1:20}".format(t,stem_word))
     
```

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
fm=open("sample.txt",'r',errors='ignore')
x=fm.read()
print(x)
punc="?:!.,;"
sentences = nltk.word_tokenize(x)
for word in sentences:
    if word in punc:
        sentences.remove(word)

sentences
print("\n{0:20}{1:20}\n".format("Word","Lemma"))
for word in sentences:
    print ("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word)))
```
